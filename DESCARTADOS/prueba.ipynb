```python
import pre_pross as prpr
import grifflin
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

base=prpr.open_bdd('./LP_IGEPN_1.json')
basef=prpr.extraer_señales(base, 'VT')
prpr.normalizar_tamanio_base(basef, 83)
prpr.normalize_samplerate_100(basef)
prpr.normalizar_muestras(basef, 8192)

print(basef.to_string())
```
